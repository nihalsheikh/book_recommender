```python
# Use to take in the description from books, and change to a form which langchain can work with
from langchain_community.document_loaders import TextLoader

# To break the description into meaningful chunks
from langchain_text_splitters import CharacterTextSplitter
from langchain_core.documents import Document

# Using OpenAI required credits for API usage. To convert the chunks in document embeddings
# from langchain_openai import  OpenAIEmbeddings

# Using Google Gemini Model: To convert the chunks in document embeddings. Alternative to OpenAI
from langchain_google_genai import GoogleGenerativeAIEmbeddings

# Using huggingface Model: To convert the chunks in document embeddings. Alternative to OpenAI
from langchain_huggingface import HuggingFaceEmbeddings

import os

# To store chunks into vector database
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
raw_documents = TextLoader("tagged_description.txt").load()

# Following Code line gave ValueError chunks must be  > 0, got 0 because of langchain package
# text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
# new line instead
lines = raw_documents[0].page_content.split("\n")

# Create a Document for each book description
documents = [Document(page_content=line.strip()) for line in lines if line.strip()]
```

```python
documents[0]
```

```python
# Not working as OpenAI requires credit and Google Gemini hits free tier limit very quickly
# db_books = Chroma.from_documents(documents, embedding=GoogleGenerativeAIEmbeddings(model="models/embedding-001"))

# Create HuggingFace embeddings (completely free, no API key!)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

db_books = Chroma.from_documents(documents, embedding=embeddings)
```

```python
# checking if the recommendation system works
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
# return user the book title from recommendation system we made
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
# Earlier we checked if recommendation query return a title for the book, now we make a func to perform the same action whenever required
def retrieve_sematic_recommendation(query: str, top_k: int = 10) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_sematic_recommendation("A book to teach children about nature")
```
